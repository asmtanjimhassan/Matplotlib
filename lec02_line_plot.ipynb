```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize = (16,10), dpi = 200)

plt.style.use('fivethirtyeight')
data = pd.read_csv('Datasets/lec01_03_data.csv')
Age = data['Age']
All_Devs = data['All_Devs']
Python = data['Python']
JavaScript = data['JavaScript']


plt.plot(Age, Python, label = 'Python')
plt.plot(Age, All_Devs, label = "All Developers")
plt.plot(Age, JavaScript, label = 'JavaScript')

plt.xlabel("Ages")
plt.ylabel("Salary")

median = 60000



plt.fill_between(Age, Python, median,
                 where = (Python > median),
                 interpolate = True,
                 alpha = 0.2)

plt.fill_between(Age, Python, median,
                 where = (Python < median),
                 interpolate = True,
                 color = 'red',
                 alpha = 1,
                label = 'Below Avg')

plt.legend()
plt.tight_layout()

plt.savefig('Reports/lec02_line_plot.png', dpi=300)
plt.show()
```
