```python
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

plt.figure(figsize = (15,5), dpi = 300)
plt.style.use('fivethirtyeight')
data = pd.read_csv('Suicide_Database_2023.csv')

desired_data = data['Reason for committing suicide']

cause_and_cases = Counter(desired_data).most_common()

cause =[]
cases =[]

for item in cause_and_cases:
    cause.append(item[0])
    cases.append(item[1])

plt.barh(cause, cases
        )
plt.ylabel("Cause of Committing Suicide")
plt.xlabel("Number of Cases Found")
plt.tight_layout()
plt.savefig('Reports/Cause of Suicide - 2.png', dpi = 300)


plt.show()
```
