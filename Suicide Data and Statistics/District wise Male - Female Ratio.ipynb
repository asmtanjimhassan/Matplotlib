```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

plt.style.use('ggplot')
plt.figure(figsize = (3,3), dpi = 300)

data = pd.read_csv('Suicide_Database_2023.csv')

district = data['District']
gender = data['Gender']

district = np.array(district)
gender = np.array(gender)

# print(counter(district))
district_desired = []
gender_desired = []

for i in range(len(district)):
    if district[i] == 'Dhaka':
        gender_desired.append(gender[i])

counting_gender = Counter(gender_desired).most_common()

gender = []
cases = []
for i in counting_gender:
    gender.append(i[0])
    cases.append(i[1])

explode = [0.05, 0.05]
plt.pie(cases,
        labels = gender,
        explode = explode,
        autopct = '%1.1f%%',
        textprops = {'fontsize' : 6}
       )
plt.title("Male Female Ratio in Bogura", fontdict={'fontsize' : 6})
plt.savefig('Reports/Male Female Ratio/Bogura', dpi = 300)

plt.show()
```
