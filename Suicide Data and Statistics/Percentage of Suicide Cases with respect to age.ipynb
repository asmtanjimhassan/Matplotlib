```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from collections import Counter

plt.style.use('ggplot')
plt.figure(figsize = (3,2), dpi = 300)
plt.rcParams['font.size'] = 3

data = pd.read_csv("Suicide_Database_2023.csv")


age = data['Age Range']

counting = Counter(age).most_common(5)

age_range = []
number = []

for items in counting:
    age_range.append(items[0])
    number.append(items[1])

expl = [0, 0.1, 0, 0]
color = ['#F9E79F', '#F5CBA7', '#E67E22', '#F5B041']

plt.pie(number,
        labels = age_range,
        startangle=90,
        colors = color,
        wedgeprops = {'edgecolor' : 'black', 'linewidth': 0.2},
        autopct='%1.1f%%'
       )

plt.title("Percentage of Suicide Cases with respect to age", fontdict={'fontsize': 6})
plt.savefig('piechart4.png', dpi = 500)
plt.show()
```
