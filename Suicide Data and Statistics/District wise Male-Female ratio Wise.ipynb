```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

plt.style.use('ggplot')
plt.figure(figsize = (3,3), dpi = 300)

data = pd.read_csv('Suicide_Database_2023.csv')

district = data['District']
edu = data['Educational Institution']

district = np.array(district)
edu = np.array(edu)

print(Counter(district))
edu_desired = []

for i in range(len(district)):
    if district[i] == 'Mymensingh':
        edu_desired.append(edu[i])

counting_category = Counter(edu_desired).most_common()

category = []
cases = []
for i in counting_category:
    category.append(i[0])
    cases.append(i[1])

explode = [0.02, 0.02, 0.02, 0.02]
plt.pie(cases,
        labels = category,
        autopct = '%1.1f%%',
        textprops = {'fontsize' : 6}
       )
plt.title("Ratio in Educational Institution in Chittagong", fontdict={'fontsize' : 6})
plt.tight_layout()
# plt.savefig('Reports/Dhaka/Educational Institution in Chittagong', dpi = 300)


plt.show()
```

```python
way_of_death = data['Way of committing suicide']
way_of_death = np.array(way_of_death)
way_of_death_desired = []

for i in range(len(district)):
    if district[i] == 'Bogura':
        way_of_death_desired.append(way_of_death[i])

counting_category = Counter(way_of_death_desired).most_common()

category = []
cases = []
for i in counting_category:
    category.append(i[0])
    cases.append(i[1])

explode = [0.02, 0.02]
plt.pie(cases,
        labels = category,
        explode = explode,
        autopct = '%1.1f%%',
        textprops = {'fontsize' : 6}
       )
plt.title("Way of Committing Suicide only in Bogura", fontdict={'fontsize' : 6})
plt.tight_layout()
#plt.savefig('Reports/Dhaka/Way of Committing Suicide only in Bogura', dpi = 300)


plt.show()
```
