```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

plt.style.use('ggplot')
plt.figure(figsize = (6,6), dpi = 300)

data = pd.read_csv('Suicide_Database_2023.csv')

district = data['District']
district = np.array(district)

age = data['Age Range']
age = np.array(age)
age_desired = []

for i in range(len(district)):
    if district[i] == 'Bogura':
        age_desired.append(age[i])

counting_category = Counter(age_desired).most_common()

category = []
cases = []
for i in counting_category:
    category.append(i[0])
    cases.append(i[1])
explode = []
for i in range(len(category)):
    explode.append(0.02)

plt.pie(cases,
        labels = category,
        explode = explode,
        autopct = '%1.1f%%',
        wedgeprops = {'linewidth' : 0.1},
        textprops = {'fontsize' : 6}
       )
plt.title("Age Range in Bogura", fontdict={'fontsize' : 12})
plt.tight_layout()
plt.savefig('Reports/Age Range/Age Range in Bogura', dpi = 300)


plt.show()
```
