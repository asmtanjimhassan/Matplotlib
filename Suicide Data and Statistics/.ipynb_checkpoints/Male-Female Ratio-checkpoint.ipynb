```python
import numpy as np
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

plt.figure(figsize = (3,3), dpi = 300)
plt.style.use('ggplot')
data = pd.read_csv('Suicide_Database_2023.csv')
desired_data = data['Gender']

male_female = Counter(desired_data).most_common()

gender =[]
cases =[]

for item in male_female:
    gender.append(item[0])
    cases.append(item[1])

color = ['red', 'orange']
explode = [0.05, 0.05]
plt.pie(cases,
        labels = gender,
        colors = color,
        explode = explode,
        autopct = '%1.1f%%',
        textprops = {'fontsize' : 6}
       )
plt.title("Male Female Ratio", fontdict={'fontsize' : 6})
plt.savefig('Reports/male-female ratio - 1.png', dpi = 300)

plt.show()
```

```python
plt.bar(gender, cases, width = 0.8)
plt.savefig('Reports/Male Female Ratio - 2.png', dpi = 300)
plt.show()
```
