```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize=(10,10), dpi=300)

data = pd.read_csv('Datasets/lec01_03_data.csv')
ages = data['Age']

bins = [0, 10,20,30,40,50, 60, 70]

plt.hist(ages,
         bins = bins,
         edgecolor = 'black',
        log = True)

median = 29

plt.axvline(
    median,
    linewidth = 2,
    color = 'green',
    label = "Median Age"
)

plt.legend()
plt.tight_layout()

plt.savefig('Reports/lec03_histogram.png', dpi=300)
plt.show()
```
