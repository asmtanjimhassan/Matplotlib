```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('ggplot')

data = pd.read_csv('Datasets/lec04_05_data.csv')
ids = data['Responder_id']
language = data['LanguagesWorkedWith']

lan_counter = Counter()
for item in language:
    my_list = item.split(';')
    lan_counter.update(my_list)

languages = []
popularity = []

for item in lan_counter.most_common(5):
    languages.append(item[0])
    popularity.append(item[1])
    
explode = [0, 0, 0, 0.1, 0]
plt.pie(popularity, labels = languages, explode = explode, startangle = 60,
        shadow = True, wedgeprops = {'edgecolor' : 'black'},
       autopct='%1.1f%%')

plt.savefig('Reports/lec05_piechart.png', dpi =400)
plt.show()
```
