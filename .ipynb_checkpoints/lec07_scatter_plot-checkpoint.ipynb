```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize = (8,5), dpi = 300)
plt.style.use('ggplot')

data = pd.read_csv('Datasets/lec07_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes,
            c = ratio,
            cmap ='summer',
            s=10,
            edgecolor = 'black',
            linewidth = 0.5,
            alpha = 0.6,
           )
plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label("Like/Dislike Ratio")

plt.xlabel("View Count")
plt.ylabel("Ratio")

plt.tight_layout()
plt.savefig('Reports/lec07_scatter_plot.png', dpi=300)
plt.show()
```
