```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


data = pd.read_csv('Datasets/lec01_03_data.csv')
Age = data['Age']
All_Devs = data['All_Devs']
Python = data['Python']
JavaScript = data['JavaScript']

fig, (ax1, ax2) = plt.subplots(nrows = 2, ncols =1, figsize=(12,18))

fig.subplots_adjust(wspace=1, hspace=5)

ax1.plot(Age, Python, label = 'Python')
ax1.plot(Age, JavaScript, label = 'JavaScript')

ax2.plot(Age, All_Devs, label = "All Developers")

ax1.legend()
ax1.set_title('Salary of Python and JS Developers vs Ages')
ax1.set_xlabel("Ages")
ax1.set_ylabel("Salary")

ax2.legend()
ax2.set_title('Salary of All Developers vs Ages')
ax2.set_xlabel("Ages")
ax2.set_ylabel("Salary")

plt.tight_layout()

plt.savefig('Reports/lec09_subplot.png', dpi=300)
plt.show()
```
